```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python

laplacian = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

image = plt.imread('assets/voiture_color.png')

filtered_image = np.zeros(image.shape)
for i in range(image.shape[2]):
    filtered_image[:, :, i] = signal.convolve2d(
        image[:, :, i], laplacian, mode='same')

enhanced_image = np.clip(image - filtered_image, 0, 255)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image)
axes[0].set_title('Image originale')
axes[1].imshow(filtered_image)
axes[1].set_title('Image filtrée avec un masque Laplacien')
axes[2].imshow(enhanced_image)
axes[2].set_title('Image rehaussée avec égalisation d\'histogramme')

for ax in axes:
    ax.set_axis_off()

plt.savefig('assets/voiture_laplacien.jpg')
```
