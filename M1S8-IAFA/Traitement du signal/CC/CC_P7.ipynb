```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

laplacian = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

image = plt.imread('assets/drapeau_Italie.jpg')

final_image = image.copy()

masque = (final_image[:, :, 1] > 100)
masque = (final_image[:, :, 0] < 100)

imgf = np.zeros(final_image.shape, dtype=np.uint8)
imgf[masque, 0] = final_image[masque, 0]
imgf[masque, 1] = final_image[masque, 1]
imgf[masque, 2] = final_image[masque, 2]
only_green_image = imgf

new_drapeau = image.copy()

mask = np.any(only_green_image != [0, 0, 0], axis=-1)

dim = len(new_drapeau[mask, 0])

new_drapeau[mask, 0] = np.full(dim, 0)
new_drapeau[mask, 1] = np.full(dim, 85)
new_drapeau[mask, 2] = np.full(dim, 164)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image)
axes[0].set_title('Image originale')
axes[1].imshow(only_green_image)
axes[1].set_title('Image filtrée')
axes[2].imshow(new_drapeau)
axes[2].set_title('Image changée')

for ax in axes:
    ax.set_axis_off()

# plt.show()
plt.savefig('assets/drapeau_Italie_green.jpg')
```
